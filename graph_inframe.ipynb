```python
import seaborn as sns
import pandas as pd
import tkinter as tk
from tkinter import *
from tkinter.ttk import *
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
matplotlib.use('TkAgg')
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
root = Tk()
root.geometry("300x300")
def graph():
    data = pd.read_csv("data.csv")
    val = drop_down.get()
    if(val == "Target"):
        data1 = data["target"]
        fig = plt.Figure(figsize = (6,6))
        a = fig.add_subplot(111)
        sns.countplot(data1)
        plt.title("Graph of Target")
        can = FigureCanvasTkAgg(fig,root)
        can.get_tk_widget()
        can.draw()
    elif(val == "fbs"):
        data1 = data["fbs"]
        fig = plt.Figure(figsize = (6,6))
        a = fig.add_subplot(111)
        sns.countplot(data1)
        plt.title("Graph of fbs")
        can = FigureCanvasTkAgg(fig,root)
        can.get_tk_widget()
        can.draw()
    elif(val == "ca"):
        data1 = data["ca"]
        fig = plt.Figure(figsize = (6,6))
        a = fig.add_subplot(111)
        sns.countplot(data1)
        plt.title("Graph of ca")
        can = FigureCanvasTkAgg(fig,root)
        can.get_tk_widget()
        can.draw()
    elif(val == "restecg"):
        data1 = data["restecg"]
        fig = plt.Figure(figsize = (6,6))
        a = fig.add_subplot(111)
        sns.countplot(data1)
        plt.title("Graph of restecg")
        can = FigureCanvasTkAgg(fig,root)
        can.get_tk_widget()
        can.draw()
    elif(val == "exang"):
        data1 = data["exang"]
        fig = plt.Figure(figsize = (6,6))
        a = fig.add_subplot(111)
        sns.countplot(data1)
        plt.title("Graph of exang")
        can = FigureCanvasTkAgg(fig,root)
        can.get_tk_widget()
        can.draw()
    elif(val == "slope"):
        data1 = data["slope"]
        fig = plt.Figure(figsize = (6,6))
        a = fig.add_subplot(111)
        sns.countplot(data1)
        plt.title("Graph of slope")
        can = FigureCanvasTkAgg(fig,root)
        can.get_tk_widget()
        can.draw()
    elif(val == "thal"):
        data1 = data["thal"]
        fig = plt.Figure(figsize = (6,6))
        a = fig.add_subplot(111)
        sns.countplot(data1)
        plt.title("Graph of thal")
        can = FigureCanvasTkAgg(fig,root)
        can.get_tk_widget()
        can.draw()
    elif(val == "cp"):
        data1 = data["cp"]
        fig = plt.Figure(figsize = (6,6))
        a = fig.add_subplot(111)
        sns.countplot(data1)
        plt.title("Graph of cp")
        can = FigureCanvasTkAgg(fig,root)
        can.get_tk_widget()
        can.draw()
drop_down = Combobox(root)
drop_down["values"] = ("Target" , "fbs", "ca", "restecg", "exang", "slope", "thal", "cp")
drop_down.current(0)
drop_down.grid(column = 0, row = 0)
but = Button(root, text = "Show Graph", command = graph)
but.grid(column = 0, row = 2)
root.mainloop()
```
