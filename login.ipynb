```python
import seaborn as sns
import pandas as pd
import tkinter as tk
from tkinter import *
#from tkinter.ttk import *
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
matplotlib.use('TkAgg')
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
data = pd.read_csv("data.csv")
root = Tk()
root.configure(background = "#333131")
root.geometry("400x400")
lbl1 = Label(root, text = "HEART DISEASE PREDICTION", font = 10, fg = "#8A6136", bg = '#333131')
lbl1.place(relx = 0.5, rely = 0.05, anchor = CENTER)  
lbl2 = Label(root, text = "User name", font = 7, fg = 'white', bg = '#333131')
lbl2.place(relx = 0.4, rely = 0.3, anchor = "center")
lbl3 = Label(root, text = "Password", font = 7, fg = 'white', bg = '#333131')
lbl3.place(relx = 0.4, rely = 0.4, anchor = "center")
txt1 = Entry(root)
txt1.place(relx = 0.55, rely = 0.3, anchor = "center")
txt2 = Entry(root)
txt2.place(relx = 0.55, rely = 0.4, anchor = "center")

def register():
    frame_2 = Tk()
    frame_2.configure(background = "#633402")
    frame_2.geometry("400x400")
    lbl2 = Label(frame_2, text = "User name", font = 7, fg = 'white', bg = '#333131')
    lbl2.place(relx = 0.4, rely = 0.3, anchor = "center")
    lbl3 = Label(frame_2, text = "Password", font = 7, fg = 'white', bg = '#333131')
    lbl3.place(relx = 0.4, rely = 0.4, anchor = "center")
    txt1 = Entry(frame_2)
    txt1.place(relx = 0.6, rely = 0.3, anchor = "center")
    txt2 = Entry(frame_2)
    txt2.place(relx = 0.6, rely = 0.4, anchor = "center")
    register_btn = Button(frame_2, text = "Register", command = frame_2.destroy, foreground = 'red', background = 'black', font = 14)
    register_btn.place(relx = 0.5, rely = 0.7, anchor = "center")
    frame_2.mainloop()

def target():
    data1 = data["target"]
    fig = plt.Figure(figsize = (6,6))
    a = fig.add_subplot(111)
    sns.countplot(data1)
    plt.title("Graph of Target")
    can = FigureCanvasTkAgg(fig,root)
    can.get_tk_widget()
    can.draw()

def fbs():
    data1 = data["fbs"]
    fig = plt.Figure(figsize = (6,6))
    a = fig.add_subplot(111)
    sns.countplot(data1)
    plt.title("Graph of fbs")
    can = FigureCanvasTkAgg(fig,root)
    can.get_tk_widget()
    can.draw()

def ca():
    data1 = data["ca"]
    fig = plt.Figure(figsize = (6,6))
    a = fig.add_subplot(111)
    sns.countplot(data1)
    plt.title("Graph of ca")
    can = FigureCanvasTkAgg(fig,root)
    can.get_tk_widget()
    can.draw()

def restecg():
    data1 = data["restecg"]
    fig = plt.Figure(figsize = (6,6))
    a = fig.add_subplot(111)
    sns.countplot(data1)
    plt.title("Graph of restecg")
    can = FigureCanvasTkAgg(fig,root)
    can.get_tk_widget()
    can.draw()

def exang():
    data1 = data["exang"]
    fig = plt.Figure(figsize = (6,6))
    a = fig.add_subplot(111)
    sns.countplot(data1)
    plt.title("Graph of exang")
    can = FigureCanvasTkAgg(fig,root)
    can.get_tk_widget()
    can.draw()

def slope():
    data1 = data["slope"]
    fig = plt.Figure(figsize = (6,6))
    a = fig.add_subplot(111)
    sns.countplot(data1)
    plt.title("Graph of slope")
    can = FigureCanvasTkAgg(fig,root)
    can.get_tk_widget()
    can.draw()

def thal():
    data1 = data["thal"]
    fig = plt.Figure(figsize = (6,6))
    a = fig.add_subplot(111)
    sns.countplot(data1)
    plt.title("Graph of thal")
    can = FigureCanvasTkAgg(fig,root)
    can.get_tk_widget()
    can.draw()

def cp():
    data1 = data["cp"]
    fig = plt.Figure(figsize = (6,6))
    a = fig.add_subplot(111)
    sns.countplot(data1)
    plt.title("Graph of cp")
    can = FigureCanvasTkAgg(fig,root)
    can.get_tk_widget()
    can.draw()
    
def userlogin():
    
    def predict():
        prediction = data.drop("target", axis = 1)
        target = data["target"]
        X_train, X_test, Y_train, Y_test = train_test_split(prediction, target, test_size = 0.25, random_state = 0)
        knn = KNeighborsClassifier(n_neighbors = 7)
        knn.fit(X_train, Y_train)
        predict_knn = knn.predict(X_test)
        score_knn = round(accuracy_score(predict_knn, Y_test)*100, 2)
        ans = Label(frame_3, text = score_knn, font = 20, fg = "white", bg = 'black')
        ans.place(relx = 0.55, rely = 0.9, anchor = "center")
    
    frame_3 = Tk()
    frame_3.geometry("400x400")
    frame_3.configure(background = "black")
    lbl1 = Label(frame_3, text = "Select The Graph", font = 20, fg = "white", bg = 'black')
    lbl1.place(relx = 0.5, rely = 0.05, anchor = CENTER)  
    target_btn = Button(frame_3, text = "Target ", command = target, foreground = "red", background = "black" , font = 10)
    fbs_btn = Button(frame_3, text = "Fbs    ", command = fbs, foreground = "red", background = "black" , font = 10)
    ca_btn = Button(frame_3, text = "ca      ", command = ca, foreground = "red", background = "black" , font = 10)
    restecg_btn = Button(frame_3, text = "restecg", command = restecg, foreground = "red", background = "black" , font = 10)
    exang_btn = Button(frame_3, text = "exang  ", command = exang, foreground = "red", background = "black" , font = 10)
    slope_btn = Button(frame_3, text = "Slope  ", command = slope, foreground = "red", background = "black" , font = 10)
    thal_btn = Button(frame_3, text = "Thal   ", command = thal, foreground = "red", background = "black" , font = 10)
    cp_btn = Button(frame_3, text = "cp     ", command = cp, foreground = "red", background = "black" , font = 10)
    predict_btn = Button(frame_3, text = "Predict", command = predict, foreground = "red", background = "black" , font = 10)
    predict_btn.place(relx = 0.45, rely = 0.9, anchor = "center")
    target_btn.place(relx = 0.25, rely = 0.15, anchor = "center")
    fbs_btn.place(relx = 0.25, rely = 0.35, anchor = "center")
    ca_btn.place(relx = 0.25, rely = 0.55, anchor = "center")
    restecg_btn.place(relx = 0.25, rely = 0.75, anchor = "center")
    exang_btn.place(relx = 0.75, rely = 0.15, anchor = "center")
    slope_btn.place(relx = 0.75, rely = 0.35, anchor = "center")
    thal_btn.place(relx = 0.75, rely = 0.55, anchor = "center")
    cp_btn.place(relx = 0.75, rely = 0.75, anchor = "center")
    exit_btn = Button(frame_3, text = "EXIT", command = frame_3.destroy, foreground = "red", background = "black" , font = 10)
    exit_btn.place(relx = 0.9, rely = 0.9, anchor = "ne")
    frame_3.mainloop()
    
new_user = Button(root, text = "New user", command = register, foreground = 'red', background = "black", font = 10)
new_user.place(relx = 0.4, rely = 0.7, anchor = 'center')
login = Button(root, text = "Login", command = userlogin, foreground = 'red', background = "black", font = 10)
login.place(relx = 0.6, rely = 0.7, anchor = 'center')
root.mainloop()
```
